## Lectura de datos en Python

### Leer datos en texto plano "manualmente"

```python
datos = []
with open("sunspots.dat") as f:
  line = f.readline()
  for line in f:
    x, y = [float(d) for d in line.split()]
    datos.append((x, y))  
```

```python
print(datos)
```

### Leyendo un CSV liberar√≠a nativa csv

```python
import csv  
datos = []
with open("CO2.csv") as f:
  reader = csv.reader(f)
  #next(reader)
  for row in reader:
    mes = float(row[0])
    CO2 = float(row[1])
    datos.append((mes, CO2))
```

```python
print(datos)
```
