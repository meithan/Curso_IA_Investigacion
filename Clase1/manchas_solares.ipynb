```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar los datos
data = np.loadtxt("sunspots.dat")
time = data[:, 0]
counts = data[:, 1]
```

```python
# Graficar
plt.figure(figsize=(12,5))
plt.plot(time, counts, lw=1)
plt.ylabel("Nmero manchas solares")
plt.title("Conteo diario de manchas solares")
plt.grid(ls=":")
plt.tight_layout()
plt.show()
```

```python
# Calcular la transformada de Fourier
power = np.abs(np.fft.rfft(counts))**2
freqs = np.fft.rfftfreq(len(counts))
```

```python
# Quitar trmino DC
power = power[1:]
freqs = freqs[1:]
periods = 1/freqs/365
```

```python
# Graficar espectro de potencia
plt.figure(figsize=(8,5))
plt.loglog(periods, power)
plt.gca().invert_xaxis()
plt.xlabel("Period (years)")
plt.ylabel("Power")
plt.title("Espectro de potencia de Fourier de manchas solares")
plt.grid(True)
#plt.axvline(11, color="red")
plt.tight_layout()
plt.show()
```
